## Reading Stock Information for Amazon using Panda's Reader

```python
#First, we have to install the packages
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#First, we need to define the start/end dates
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)
```

```python
#Next we can read the data from amazon
AMZN = data.DataReader("AMZN", 'yahoo', start, end)
```

```python
AMZN
```

```python
#We need to reset the index
myAMZN = AMZN.reset_index()
```

```python
myAMZN
```

### Creating a a time series (line chart) of the volume from January 1, to today

```python
plt.style.use('ggplot')
myAMZN.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```

### Plotting a bar chart of the "opening" for the month of December 2021

```python
#We need to change our time frame
start_2 = datetime.datetime(2021, 12, 1)
end_2 = datetime.datetime(2021, 12, 31)
```

```python
MonthAMZN = data.DataReader("AMZN", 'yahoo', start_2, end_2)
```

```python
MonthAMZN = MonthAMZN.reset_index()
```

```python
MonthAMZN.plot.bar(x='Date',y = 'Open')
```
